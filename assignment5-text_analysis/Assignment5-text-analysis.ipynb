```python
def my_tokenizer(txt):
    return txt.split(' ')
```

```python
def remove_punctuations_and_numbers(text):
    # Remove punctuations and numbers using regular expression
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\s+', ' ', text)
    text = re.sub(r'[\W\d]+', ' ', text)
    text = text.strip()
    
    return text
```

```python
df_train_clean = df_train.copy()
# cleaning the training set
for i in range(len(df_train_clean)):
    df_train_clean.iloc[i]['story'] = remove_punctuations_and_numbers(df_train_clean.iloc[i]['story'])
df_train_clean.iloc[2]['story']
df_train.iloc[2]['story']
```

```python

# tokenizing the text 
vec = CountVectorizer(tokenizer=my_tokenizer)
vec = TfidfVectorizer(use_idf=False,sublinear_tf=True,min_df=5,max_df=450,max_features=1000)
X = vec.fit_transform(df_train_clean['story'])
pd.DataFrame(X.toarray(), columns=vec.get_feature_names_out())
```

```python
# split data into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, df_train_clean['gender'], test_size=0.2, random_state=42)
# Train a Naive Bayes classifier
clf_NB = MultinomialNB()
clf_NB.fit(X_train, y_train)
# Train a KNN classifier
clf_KNN = KNeighborsClassifier(n_neighbors=5)
clf_KNN.fit(X_train, y_train)
# Train a Decision tree classifier
clf_Dt = DecisionTreeClassifier(max_depth=7)
clf_Dt.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report

# Predict the labels for the test data using each classifier
y_pred_NB = clf_NB.predict(X_test)
y_pred_KNN = clf_KNN.predict(X_test)
y_pred_Dt = clf_Dt.predict(X_test)

# Generate the classification report for each classifier
report_NB = classification_report(y_test, y_pred_NB, target_names=['female', 'male'])
report_KNN = classification_report(y_test, y_pred_KNN, target_names=['female', 'male'])
report_Dt = classification_report(y_test, y_pred_Dt, target_names=['female', 'male'])

# Print the classification reports
print("Naive Bayes Classification Report:")
print(report_NB)
print('cross validation: ', cross_val_score(clf_NB, X_test, y_test, cv=5, scoring='f1_macro').mean())
print()
print("KNN Classification Report:")
print(report_KNN)
print('cross validation: ', cross_val_score(clf_KNN, X_test, y_test, cv=5, scoring='f1_macro').mean())
print()
print("Decision Tree Classification Report:")
print(report_Dt)
print('cross validation: ', cross_val_score(clf_Dt, X_test, y_test, cv=5, scoring='f1_macro').mean())
```
